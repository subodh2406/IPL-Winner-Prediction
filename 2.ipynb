```python
# Dataset 1 (Match Information)
# This dataset contains match-level information, including details such as:

#         id: Unique match identifier.
#         Season: IPL season.
#         City: City where the match took place.
#         Date: Date of the match.
#         Team1 and Team2: The teams playing the match.
#         Toss Winner and Toss Decision: Details of the team that won the toss and their decision to bat or field.
#         Result & Winner: Whether the result was normal or affected by D/L method and the winning team.
#         Win by Runs/Win by Wickets: How the match was won (either by runs or wickets).
#         Player of the Match: The best performer of the match.
#         Venue: The stadium where the match took place.
#         Umpires: Umpires overseeing the match.



# Dataset 2 (Ball-by-Ball Information)
# This dataset gives ball-by-ball details for each match:

#         match_id: The unique match ID (links with the first dataset).
#         inning: The inning number (1st or 2nd).
#         batting_team, bowling_team: Teams involved in batting and bowling for that delivery.
#         over, ball: Over and ball number for the specific delivery.
#         batsman, non_striker, bowler: Players involved in that delivery (batsman on strike, non-striker, and bowler).
#         Various run types: Fields like wide_runs, legbye_runs, etc., describe how the runs were accumulated in that delivery.
#         batsman_runs, total_runs: Runs scored by the batsman on that delivery and the overall runs (including extras).
#         player_dismissed, dismissal_kind, fielder: Information about the dismissal (if a wicket fell), the type of dismissal (bowled, caught, etc.), and the fielder involved.
        
# Dataset Summary:
# Matches Dataset (matches.csv)
# Rows: 756 (Each row represents a match)
# Columns: 18 (Details like teams, venue, winner, etc.)

# Deliveries Dataset (deliveries.csv)
# Rows: 179,078 (Each row represents a ball bowled)
# Columns: 21 (Details like runs, bowler, batsman, dismissals, etc.)
```

```python
# total run chahiye match jitne ke liye 
# matlab runs score by team in 1st inning
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
total_score_df.head()
```

```python
# apne ko sirf 1st inning ke runs cahiye

total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
# do data frame ko merge kar arha ahi math aur total_score on basic of match id
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# purane nam naye nam se change kar rahe hai
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
# retired teams ko nukal rahe hai
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
# Duckworth-Lewis method
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
# mergin delivering dataset with the matches dataset now we have ball by balls record of each match
delivery=match_df.merge(delivery,on='match_id')
delivery
```

```python
# sirf inning ke upper kam kar rahe hai isslye matlab chase karte waqt data chahiye iss liye recird rakh erahe  hai 
delivery_df = delivery[delivery['inning']==2]
```

```python
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
# yaha se running total score current score nikal rahe hai matlab har ek ball ke bad kitne runs bane
# total_runs_y ek feature hai jo bata hai uss bal pe kitne runs bane hai 
```

```python
# runs left nikal rahe hai har ek ball bad
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
# balls left
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
# Fill missing values and convert 'player_dismissed' to binary (0 or 1)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == "0" else 1).astype(int)

# Calculate cumulative sum of player_dismissed for wickets
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()

# Calculate wickets remaining
delivery_df['wickets'] = 10 - wickets

# Ensure 'wickets' doesn't go below 0
delivery_df['wickets'] = delivery_df['wickets'].clip(lower=0)

# Display the updated DataFrame
print(delivery_df.head())



# yaha se kitne wickets bache hai vho nhikal rahe hai phele player dissmissed ka nam tha usko 0 ya 1 me convvert kar rareh hai phir calculate 10 se minus 
# kar rahe hai aur remiainig wickets nikal rahe hai 
```

```python
# crr = runs/overs
# It is calculated by taking the total runs scored so far (current score) and dividing it by the number of overs (balls) faced.
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# It is calculated by dividing the remaining runs by the number of balls left and then multiplying by 6 to get the rate per over.
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# result find kar rahe hai # winning team aur batting aga same hai toh one aa jayega 

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```
